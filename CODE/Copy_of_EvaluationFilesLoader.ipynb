```python
# Import necessary libraries
import os
import pandas as pd
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Path to your folder containing the CSV files
folder_path = '/content/drive/MyDrive/TFM2/TFM-DATASETS/evaluations'

# Path where the combined CSV will be saved
output_file = '/content/drive/MyDrive/TFM2/TFM-DATASETS/evaluations_combined.csv'

# Read all CSV files and combine them
all_data = []
for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):
        file_path = os.path.join(folder_path, filename)
        print(f"Loading {filename}...")

        # Load the CSV into a Pandas DataFrame
        df = pd.read_csv(file_path)

        # Add a column to track the source file
        df['source_file'] = filename

        all_data.append(df)

# Concatenate all DataFrames into one
df_combined = pd.concat(all_data, ignore_index=True)

# Save the combined DataFrame to a single CSV file
df_combined.to_csv(output_file, index=False)

print(f"Combined CSV saved to: {output_file}")

```
